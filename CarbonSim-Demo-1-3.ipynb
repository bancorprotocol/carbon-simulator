# Carbon Simulation - Demo 1-3

In this demo we look at **multi-curve unidirectional liquidity with a zero width range**

```python
Sim = CarbonSimulatorUI(pair="ETHUSDC", verbose=False)
Sim
```

We are setting up unidirectional ranges for selling ETH at 2000, 2010, ..., 2040 and seed them with 10 ETH each

```python
Sim.add_order("ETH", 10, 2000, 2000)
Sim.add_order("ETH", 10, 2010, 2010)
Sim.add_order("ETH", 10, 2020, 2020)
Sim.add_order("ETH", 10, 2030, 2030)
Sim.add_order("ETH", 10, 2040, 2040)
Sim.state()["orders"]
```

We are now selling five times 10 ETH and we see that they indeed just match those positions

```python
Sim.amm_sells("ETH", 10)
Sim.amm_sells("ETH", 10)
Sim.amm_sells("ETH", 10)
Sim.amm_sells("ETH", 10)
Sim.amm_sells("ETH", 10)
Sim.state()["trades"].query("aggr==True")
```

All orders are no used up...

```python
Sim.state()["orders"]
```

...so we can't even sell epsilon

```python
Sim.amm_sells("ETH", 0.00001)
```
