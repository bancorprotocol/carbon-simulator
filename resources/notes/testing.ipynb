```python
# from testing_example_SHIBUSDC_shib_side import order_inputs
# from testing_example_SHIBUSDC_usdc_side import order_inputs

# from testing_example_WBTCBNT_wbtc_side import order_inputs
# from testing_example_WBTCBNT_bnt_side import order_inputs

# from testing_example_SHIBWBTC_shib_side import order_inputs
# from testing_example_SHIBWBTC_wbtc_side import order_inputs

# from testing_example_BNTDAI_bnt_side import order_inputs
from testing_example_BNTDAI_dai_side import order_inputs


resultsdf = pd.DataFrame()

### INIT ###
BITS_SIGNIFICANTs =  [48]
BITS_EXPONENTs    =   [6]
SCALING_FACTORs  =  [40, 48]

for BITS_SIGNIFICANT in BITS_SIGNIFICANTs:
    for BITS_EXPONENT in BITS_EXPONENTs:
        for SCALING_FACTOR in SCALING_FACTORs:
            storage = create_order(order_inputs, BITS_SIGNIFICANT, BITS_EXPONENT, SCALING_FACTOR)
            tradeByTarget = True
            amount_array = np.logspace(-5,log10(order_inputs['y']),int(log10(order_inputs['y'])+6))
            outputs = []
            for amount in amount_array:
                try:
                    trade_output = trade(amount, tradeByTarget, storage, order_inputs, SCALING_FACTOR)
                except AssertionError as e:
                    trade_output = "AssertionError"
                outputs += [trade_output]
            # print(outputs)

            tradeByTarget = False
            sourceOut = []
            for output in outputs:
                print("output is", output)
                if (output == 0.0) or (output == "AssertionError"):
                    sourceOut += ['Error']
                else:
                    try:
                        trade_output = trade(output, tradeByTarget, storage, order_inputs, SCALING_FACTOR)
                    except AssertionError as e:
                        trade_output = "AssertionError"
                    sourceOut += [trade_output]

            results = pd.DataFrame(list(zip(amount_array, outputs, sourceOut)), columns=['tradebyTargetInput', 'output_to_input', 'tradeBySourceOutput'])
            results['Parameters'] = [(BITS_SIGNIFICANT,BITS_EXPONENT,SCALING_FACTOR) for x in range(len(results))]
            results= results[['Parameters', 'tradebyTargetInput', 'output_to_input', 'tradeBySourceOutput']].copy()

            resultsdf = resultsdf.append(results)
```
