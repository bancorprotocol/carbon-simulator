# Carbon Simulation - Test 10

```python
Sim = CarbonSimulatorUI(verbose=False, raiseonerror=False)
Sim
```

## Trading the full position

we show that the full position, if traded in a single trade, is indeed traded at the geometric average, here 2449

```python
Sim.add_sgl_pos("ETH", 100, 2000, 3000, "ETHUSDC")
print(sqrt(2000*3000))
Sim.state()["orders"]
```

```python
Sim.amm_sells("ETH", 100, "ETHUSDC")["trades"]
```

```python
Sim.state()["orders"]
```

## Trading the position in two symmetric steps

we show that the full position, if traded in a two trades, is still effectively traded at the geometric average, here 2449

```python
Sim.add_sgl_pos("ETH", 100, 2000, 3000, "ETHUSDC")
print(sqrt(2000*3000))
Sim.state()["orders"]
```

```python
Sim.amm_sells("ETH", 50, "ETHUSDC")["trades"]
```

```python
Sim.amm_sells("ETH", 50, "ETHUSDC")["trades"]
```

```python
Sim.state()["orders"]
```

```python
Sim.state()["trades"]
```

```python
p0 = Sim.state()["trades"].iloc[0]["price_avg"]
print(f"selling 100 ETH for {p0*100} USDC (price {p0})")
```

```python
p11 = Sim.state()["trades"].iloc[2]["price_avg"]
print(f"selling 50 ETH for {p11*50} USDC (price {p11})")
```

```python
p12 = Sim.state()["trades"].iloc[4]["price_avg"]
print(f"selling 50 ETH for {p12*50} USDC (price {p12})")
```

```python
print(f"selling 100 ETH for {p11*50+p12*50} USDC (price {(p11*50+p12*50)/100})")
```
