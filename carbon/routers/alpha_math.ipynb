```python
# returns an enumerated list with only values >= val
def bigger_than(lst, val):
    return([(i,c) for i,c in lst if c >= val])

# returns the sum of the values in an enumerated list
def sum_me(lst):
    return(sum([c for i,c in lst]))

# returns the first single min value in an enumerated list
def min_item(lst):
    min_c = min([c for i,c in lst])
    min_c_list = [(i,c) for i,c in lst if c == min_c]
    if len(min_c_list) > 1:
        return(min_c_list[0])
    else:
        return(min_c_list[0])

# returns the last single min value in an enumerated list
def min_item_last(lst):
    min_c = min([c for i,c in lst])
    min_c_list = [(i,c) for i,c in lst if c == min_c]
    if len(min_c_list) > 1:
        return(min_c_list[-1])
    else:
        return(min_c_list[0])

# returns the first single max value in an enumerated list
def max_item(lst):
    max_c = max([c for i,c in lst])
    max_c_list = [(i,c) for i,c in lst if c == max_c]
    if len(max_c_list) > 1:
        return(max_c_list[0])
    else:
        return(max_c_list[0])

def get_i(lst, wanted_i):
    return([(i,c) for i,c in lst if i == wanted_i][0])

def sum_list_indexes(lst):
    return(sum([i for i,c in lst]))

# def assertion_checks(full_fill, current_sum, threshold, max_fill, threshold_list, num_values):
def assertion_checks(full_fill, current_sum, requested_trade_amount, max_fill, threshold_list, threshold_orders):
    if full_fill == True:
        assert(current_sum >= requested_trade_amount)
    else:
        assert(current_sum == max_fill)

    if len(threshold_list) < threshold_orders:
        pass
    else:
        assert(len(set([i for i,c in threshold_list])) == threshold_orders)

    assert(len(threshold_list) == len(set([i for i,c in threshold_list])))
```

```python
# returns a list of orders that meet the threshold value
def crude_threshold_orders(numbers, threshold, num_values):
    count = 0
    current_sum = 0
    indexed_values = list(enumerate(numbers))
    threshold_list = indexed_values[:num_values]
    for i in range(num_values, len(numbers)):
        count += 1
        # print(threshold_list)
        current_sum = sum_me(threshold_list)
        if current_sum >= threshold:
            # print(threshold_list)
            # print(current_sum)
            # print(threshold)
            return(threshold_list, current_sum, threshold, count)
        else:
            # print(threshold_list)
            min_index, min_val = min_item(threshold_list)
            # print(min_val)
            threshold_list.remove(min_item(threshold_list))

            threshold_list.append(indexed_values[i])
            # print(threshold_list)

    if i == len(numbers)-1:
        print('Entire list searched')
        current_sum = sum_me(threshold_list)
    if current_sum < threshold:
        print('Partial fulfillment')

    return(threshold_list, current_sum, threshold, count)

```

```python
def fast_trim_while_threshold_orders(numbers, threshold, num_values):
    count = 0
    current_sum = 0
    indexed_values = list(enumerate(numbers))
    threshold_list = indexed_values[:num_values]

    while indexed_values:
        count +=1
        # print(len(indexed_values))
        i = indexed_values[0][0]+num_values-1
        # print(i)
        current_sum = sum_me(threshold_list)
        # if this condition is met then there is enough liquidity in the selected orders to meet the trade amount
        if current_sum >= threshold:
            # print(threshold_list)
            # print(current_sum)
            # print(threshold)
            indexed_values = []
            return(threshold_list, current_sum, threshold, count)#
            # break
        else:
            # this condition ensures that we dont exceed the index val
            if i < len(numbers) - 1:
                # print(threshold_list)
                min_index, min_val = min_item(threshold_list)
                max_index, max_val = max_item(list(enumerate(numbers)))
                # this condition allows us to quit early if there are no larger amounts in the list i.e. must be partial
                if min_val == max_val:
                    return(threshold_list, current_sum, threshold, count)#
                    # break
                else:    
                    # if the next value in the list is bigger then we should drop the min and add it (might offer a 
                    # worse rate but we need to decide somewhere)
                    # print(i)
                    threshold_list.remove(min_item(threshold_list))
                    threshold_list.append(indexed_values[0])
                    # print(threshold_list)
            else:
                # if we are outside the index range then that should be it!
                return(threshold_list, current_sum, threshold, count) 
            indexed_values = bigger_than(indexed_values, min_val)[1:]
            # print(indexed_values)

    if i == len(numbers)-1:
        print('Entire list searched')
        current_sum = sum_me(threshold_list)
    if current_sum < threshold:
        print('Partial fulfillment')
    return(threshold_list, current_sum, threshold, count)#
```

```python


def partial_trim_while_threshold_orders(numbers, threshold, num_values):
    count = 0
    current_sum = 0
    indexed_values = list(enumerate(numbers))
    threshold_list = indexed_values[:num_values]
    max_fill = sum(sorted(numbers, reverse=True)[:num_values])
    full_fill = False

    if threshold <= max_fill:
        full_fill = True
        print('Full fulfillment expected')
    else:
        print(f'Partial fulfillment expected of {max_fill}')

    while (current_sum < max_fill) & (len(indexed_values) > 0):
        count +=1
        # print(len(indexed_values))
        i = indexed_values[0][0]
        # print(i)
        current_sum = sum_me(threshold_list)
        
        if (not full_fill) & (current_sum==max_fill):
            indexed_values = []
            assertion_checks(full_fill, current_sum, threshold, max_fill, threshold_list, num_values)
            return(threshold_list, current_sum, threshold, count)#

        # if this condition is met then there is enough liquidity in the selected orders to meet the trade amount
        if current_sum >= threshold:
            # print(threshold_list)
            # print(current_sum)
            # print(threshold)
            indexed_values = []
            assertion_checks(full_fill, current_sum, threshold, max_fill, threshold_list, num_values)
            return(threshold_list, current_sum, threshold, count)#
        else:
            # this condition ensures that we dont exceed the index val
            if i < len(numbers) - 1:
                # print(threshold_list)
                min_index, min_val = min_item(threshold_list)
                max_index, max_val = max_item(list(enumerate(numbers)))
                # this condition allows us to quit early if there are no larger amounts in the list i.e. must be partial
                if min_val == max_val:
                    assertion_checks(full_fill, current_sum, threshold, max_fill, threshold_list, num_values)
                    return(threshold_list, current_sum, threshold, count)#
                    
                else:    
                    # if the next value in the list is bigger then we should drop the min and add it (might offer a 
                    # worse rate but we need to decide somewhere)
                    # print(i)
                    threshold_list.remove(min_item(threshold_list))
                    threshold_list_indexes = [i for i,c in threshold_list]
                    subset_indexed_values = [(i,c) for i,c in indexed_values if i not in threshold_list_indexes]
                    threshold_list.append(subset_indexed_values[0])
                    # print(threshold_list)
            else:
                pass

            if full_fill:
                indexed_values = bigger_than(indexed_values, min_val)[1:]
            else:
                indexed_values = indexed_values[1:]
            # print(indexed_values)
```

```python
def incomplete_order_selector(numbers, threshold, num_values):
    count = 0
    current_sum = 0
    indexed_values = list(enumerate(numbers))
    threshold_list = indexed_values[:num_values]
    max_fill = sum(sorted(numbers, reverse=True)[:num_values])
    full_fill = False

    # liquidity check to determine partial fill
    if threshold <= max_fill:
        full_fill = True
        print('Full fulfillment expected')
    else:
        print(f'Partial fulfillment expected of {max_fill}')

    # at maximum, loop over the entire list less the seeded first vals
    for i,c in indexed_values[num_values:]:
        count += 1
        current_sum = sum_me(threshold_list)

        # early exit when a partial fill is met
        if (not full_fill) & (current_sum==max_fill):
            assertion_checks(full_fill, current_sum, threshold, max_fill, threshold_list, num_values)
            sum_threshold_list_indexes = sum_list_indexes(threshold_list)
            return(threshold_list, current_sum, threshold, count, sum_threshold_list_indexes)

        # early exit when we filled the order
        if current_sum >= threshold:
            assertion_checks(full_fill, current_sum, threshold, max_fill, threshold_list, num_values)
            sum_threshold_list_indexes = sum_list_indexes(threshold_list)
            return(threshold_list, current_sum, threshold, count, sum_threshold_list_indexes)
        
        # else iterate over the remaining values
        else:
            min_index, min_val = min_item(threshold_list)
            max_index, max_val = max_item(indexed_values)
            threshold_list_indexes = [i for i,c in threshold_list]
            next_val = get_i(indexed_values, i)[1]
            # if the order hasn't been filled more liquidity is needed so swap out the minimum value and take the next in the list
            # if the min_val in the threshold list is greater than the max_val in the full list things can't get any better
            if min_val >= max_val:
                assertion_checks(full_fill, current_sum, threshold, max_fill, threshold_list, num_values)
                sum_threshold_list_indexes = sum_list_indexes(threshold_list)
                return(threshold_list, current_sum, threshold, count, sum_threshold_list_indexes)       
            
            # however, the next value to insert must be greater than the current value and no double-dipping
            elif (next_val > min_val) & (i not in threshold_list_indexes):
                threshold_list.remove(min_item(threshold_list))
                threshold_list.append(get_i(indexed_values, i))
            else:
                pass
    
    # if you make it this far that means that the final order in the list is required to fufill the trade
    # should only happen in low liquidity situations
    assertion_checks(full_fill, current_sum, threshold, max_fill, threshold_list, num_values)
    sum_threshold_list_indexes = sum_list_indexes(threshold_list)
    return(threshold_list, current_sum, threshold, count, sum_threshold_list_indexes)
```

```python
def gen_one_order_selector(numbers, threshold, num_values):
    count = 0
    current_sum = 0
    indexed_values = list(enumerate(numbers))
    max_index, max_val = max_item(indexed_values)
    threshold_list = indexed_values[:num_values]
    max_fill = sum(sorted(numbers, reverse=True)[:num_values])
    full_fill = False

    # liquidity check to determine partial fill
    if threshold <= max_fill:
        full_fill = True
        # print('Full fulfillment expected')
    else:
        # print(f'Partial fulfillment expected of {max_fill}')
        pass

    # loop over the remaining values
    while indexed_values:
        count += 1
        # set i based on the the value of the first item in the list, offset by the number of values selected
        i = indexed_values[0][0]+num_values-1
        current_sum = sum_me(threshold_list)

        # early exit when a partial fill is met
        if (not full_fill) & (current_sum==max_fill):
            exit_summary = 1
            assertion_checks(full_fill, current_sum, threshold, max_fill, threshold_list, num_values)
            sum_threshold_list_indexes = sum_list_indexes(threshold_list)
            return(threshold_list, current_sum, threshold, sum_threshold_list_indexes, count, exit_summary)#

        # early exit when we filled the order
        if current_sum >= threshold:
            exit_summary = 2
            assertion_checks(full_fill, current_sum, threshold, max_fill, threshold_list, num_values)
            sum_threshold_list_indexes = sum_list_indexes(threshold_list)
            return(threshold_list, current_sum, threshold, sum_threshold_list_indexes, count, exit_summary)#
        else:
            # else iterate over the remaining values dont exceed the the total number of inputs
            if i < len(numbers) - 1:
                min_index, min_val = min_item(threshold_list)
                next_val = get_i(indexed_values, i+1)[1]

                # if the order hasn't been filled more liquidity is needed so swap out the minimum value and take the next in the list
                # if the min_val in the threshold list is greater than the max_val in the full list things can't get any better
                if min_val >= max_val:
                    exit_summary = 3
                    assertion_checks(full_fill, current_sum, threshold, max_fill, threshold_list, num_values)
                    sum_threshold_list_indexes = sum_list_indexes(threshold_list)
                    return(threshold_list, current_sum, threshold, sum_threshold_list_indexes, count, exit_summary)#
                
                # however, the next value to insert must be greater than the current value
                elif next_val > min_val:
                    threshold_list.remove(min_item(threshold_list))
                    threshold_list.append(get_i(indexed_values, i+1))
                else:
                    pass
            else:
                exit_summary = 4
                assertion_checks(full_fill, current_sum, threshold, max_fill, threshold_list, num_values)
                sum_threshold_list_indexes = sum_list_indexes(threshold_list)
                return(threshold_list, current_sum, threshold, sum_threshold_list_indexes, count, exit_summary)#
            indexed_values = indexed_values[1:]
```

```python
def gen_two_order_selector(amounts, requested_trade_amount, threshold_orders):
    '''
    GenII is an improvement on GenI both in terms of readability and performance.
    The performance comes from a minor bug fix where (in GenI) the minimum valued item in the threshold list was preferentially low index value as opposed to high index value.
    The improvement results in a lower sum_threshold_list_indexes which indicates a better price ranking.
    GenII results in a better trade rate in approximately 1% of cases.

    Amounts should have been sorted into best-price-first order

    :amounts:                       the input amounts that need to be sorted and optimized
    :requested_trade_amount:        the minimum requested_trade_amount that need to be met by sorting the amounts
    :threshold_orders:              the number of amounts that are permitted to be used in order to meet the requested_trade_amount 


    ## WHAT IS KNOWN
    1. The maximum ouput amount is limited by the threshold_orders.
    2. If the maximum output amount is less than the requested_trade_amount then this is a partial fill
    3. The maximum number of amounts that make it into the final list is n-1 occuring only when the 
        requested_trade_amount is the sum of all but the final amount in the amounts list. 
        Else the process would have finished early with a partial fill
    4. The process ends if:
        a) Partial fill is supported and the sum(amounts) < requested_trade_amount
        b) If the sum(threshold_list) >= requested_trade_amount
    5. We should not get to the point where:
        a) We have iterated through every amount and the trade has not been met, as it should have been completed with partial or insufficient liquidity

    

    ### Order of operations
    1. Check if max_fill >= requested_trade_amount to determine if the request will be fully fulfilled
    2. For each value added to the threshold list check if trade is met
    3. When adding a new amount to the list it must be greater than the minimum amount already in the list
    
    '''
    exit_summary = 0
    count = 0
    indexed_amounts = list(enumerate(amounts))
    max_index, max_val = max_item(indexed_amounts)
    threshold_list = indexed_amounts[:threshold_orders]    # set the initial threshold_list as the first orders in the list
    max_fill = sum(sorted(amounts, reverse=True)[:threshold_orders])  # identify the maximum return possible given the threshold_orders limitation
    current_sum = sum_me(threshold_list)

    # liquidity check to determine partial fill
    full_fill = True if requested_trade_amount<=max_fill else False

    # you have already populated the threshold_list so you can start counting at the next item in the amounts list
    for i in range(threshold_orders,len(amounts)+1):
        count += 1
        current_sum = sum_me(threshold_list)

        # early exit when a partial fill is met
        if (not full_fill) & (current_sum==max_fill): # this works because a partial fill requires that the requested_trade_amount is <= max_fill, and max_fill is the n biggest values in the amounts, and thus the current_sum must include the biggest
            exit_summary = 1
            assertion_checks(full_fill, current_sum, requested_trade_amount, max_fill, threshold_list, threshold_orders)
            sum_threshold_list_indexes = sum_list_indexes(threshold_list)
            return(threshold_list, current_sum, requested_trade_amount, sum_threshold_list_indexes, count, exit_summary)#

        # early exit when we filled the order
        if current_sum >= requested_trade_amount:
            exit_summary = 2
            assertion_checks(full_fill, current_sum, requested_trade_amount, max_fill, threshold_list, threshold_orders)
            sum_threshold_list_indexes = sum_list_indexes(threshold_list)
            return(threshold_list, current_sum, requested_trade_amount, sum_threshold_list_indexes, count, exit_summary)#
        else:
            min_index, min_val = min_item_last(threshold_list)
            next_val = get_i(indexed_amounts, i)[1]

            # the next value to insert must be greater than the current value
            if next_val > min_val:
                threshold_list.remove(min_item_last(threshold_list))
                threshold_list.append(get_i(indexed_amounts, i))
            else:
                pass
    exit_summary = 4
    current_sum = sum_me(threshold_list)
    assertion_checks(full_fill, current_sum, requested_trade_amount, max_fill, threshold_list, threshold_orders)
    sum_threshold_list_indexes = sum_list_indexes(threshold_list)
    return(threshold_list, current_sum, requested_trade_amount, sum_threshold_list_indexes, count, exit_summary)#
```

```python
def get_replacement_options(delta, indexed_amounts, threshold_list, current_sum, requested_trade_amount):
    threshold_list_indexes = [i for i,c in threshold_list]
    options = {}
    count = 0
    left_out_lessthan = [(i,c) for i,c in indexed_amounts if i < threshold_list_indexes[-1]]
    for out in left_out_lessthan:
        if out[1] <= delta:
            options[count] = {}
            options[count]['remove'] = []
            options[count]['append'] = []
            # print(out)
            for item in threshold_list:
                if (current_sum + out[1] - item[1] >= requested_trade_amount) & (out[0] < item[0]):
                    # print(out,":", item)
                    options[count]['append'] = out
                    options[count]['remove'] += [item]
            count+=1
    return(options)
```

```python
def evaulate_options(options):
    best_k = None
    min_result = 0
    for k in options.keys():
        list_index_improve = options[k]['append'][0] - options[k]['remove'][0]
        if list_index_improve < min_result:
            min_result = list_index_improve
            best_k = k
    return(best_k)
```

```python
def get_replacement_options_alt(delta, indexed_amounts, threshold_list, current_sum, requested_trade_amount):
    threshold_list_indexes = [i for i,c in threshold_list]
    options = {}
    count = 0
    left_out_lessthan = [(i,c) for i,c in indexed_amounts if i < threshold_list_indexes[-1]]
    for out in left_out_lessthan:
        potentials = []
        if out[1] <= delta:
            for item in threshold_list:
                if (current_sum + out[1] - item[1] >= requested_trade_amount) & (out[0] < item[0]):
                    potentials += [item] 

        if len(potentials)>0:
            options[count] = {}
            options[count]['remove'] = []
            options[count]['append'] = []
            select_i = min([i for i,c in potentials])
            selection = [(i,c) for i,c in potentials if i == select_i][0]
            options[count]['append'] = out
            options[count]['remove'] = selection
            count+=1
    
    print("options", options)
    if len(options) != 0:
        if len(options[0]['remove']) != 0:
            best_k = evaulate_options(options)
            selection = options[best_k]['append']
            if selection not in threshold_list:
                return(options[best_k])
            else:
                return(None)
        else:
            return(None)
    else:
        return(None)
```

```python
def gen_three_order_selector(amounts, requested_trade_amount, threshold_orders):
    '''
    GenIII is a layer on top of GenII that evaluates the "delta" between the current_sum and the requested_trade_amount and looks for lower index value orders that could replace
    orders in the threshold_list while keeping the delta >= 0.
    This back-screen allows orders that were dropped while trying to meet the requested_trade_amount to be revisited and added, provided the change lowers the sum_threshold_list_indexes
    and keeps delta >= 0.

    GenIII results in a better trade rate vs GenII in approximately 4% of cases.
    GenIII results in a better trade rate vs GenI in approximately 5% of cases.
    '''

    threshold_list, current_sum, requested_trade_amount, sum_threshold_list_indexes, count, exit_summary = gen_two_order_selector(amounts, requested_trade_amount, threshold_orders)
    if current_sum >= requested_trade_amount:
        full_fill = True
    else:
        full_fill = False

    indexed_amounts = list(enumerate(amounts))
    delta = current_sum - requested_trade_amount
    if delta > 0:
        # print("current_sum:",current_sum)
        # print("requested_trade_amount:",requested_trade_amount)
        # print("delta:",delta)

        options = get_replacement_options(delta, indexed_amounts, threshold_list, current_sum, requested_trade_amount)
        # print(options)

        for k in options.keys():
            delta = current_sum - requested_trade_amount
            if (delta > 0) & (len(options[k]['remove'])!=0):
                select_i = min([i for i,c in options[k]['remove']])
                selection = [(i,c) for i,c in options[k]['remove'] if i == select_i][0]
                appendme = options[k]['append']
                # print(selection)
                # print(threshold_list)
                if (selection in threshold_list) & (appendme not in threshold_list):
                    threshold_list.remove(selection)
                    threshold_list.append(appendme)
                    current_sum = sum_me(threshold_list)
                    delta = current_sum - requested_trade_amount
                    if delta < 0: # roll it back
                        threshold_list.remove(appendme)
                        threshold_list.append(selection)
                        current_sum = sum_me(threshold_list)

    sum_threshold_list_indexes = sum_list_indexes(threshold_list)
    current_sum = sum_me(threshold_list)
    
    if len(threshold_list) < threshold_orders:
        pass
    else:
        assert(len(set([i for i,c in threshold_list])) == threshold_orders)

    assert(len(threshold_list) == len(set([i for i,c in threshold_list])))

    if full_fill:
        assert(current_sum>=requested_trade_amount)

    return(threshold_list, current_sum, requested_trade_amount, sum_threshold_list_indexes, count, exit_summary)     
```
